Average customer lifetime value by type of coverage

```python


sns.barplot(x="Coverage", y="Customer Lifetime Value", data=clv_data, palette=['green', 'orange','blue'])

plt.show()
```

```python
state_clv = clv_data[["State", "Customer Lifetime Value"]].groupby("State").mean()

state_clv.index = state_clv.index.map({"Arizona": "AZ",
                                       "California": "CA",
                                       "Nevada": "NV",
                                       "Oregon": "OR",
                                       "Washington": "WA"})

fig = px.choropleth(state_clv,
                    locations=state_clv.index,
                    locationmode="USA-states",
                    color="Customer Lifetime Value",
                    scope="usa")

fig.show()

```

Average clv in each state per coverage

```python
plt.figure(figsize = [12, 8])

sns.barplot(x="State", y="Customer Lifetime Value", data=clv_data, hue="Coverage")

plt.title("State vs CLV")
plt.show()
```

```python
plt.figure(figsize = [12, 8])

sns.barplot(y="Customer Lifetime Value", hue="Number of Open Complaints", data=clv_data)


plt.show()
```

```python
plt.figure(figsize = [12, 8])

sns.barplot(y="Customer Lifetime Value", hue="Vehicle Class", data=clv_data)

plt.show()
```

Luxury suv and luxury car tend to have more CLV than others

```python
plt.figure(figsize = [12, 8])

sns.barplot(hue="Number of Policies", y="Customer Lifetime Value", data=clv_data)

plt.title("Number of Policies vs CLV")
plt.show()
```

people who buy 2 policies tend to have more CLV

```python
plt.figure(figsize = [12, 8])

sns.barplot(hue="Education", y="Customer Lifetime Value", data=clv_data)

plt.title("Education vs CLV")
plt.show()
```

Masters and high school people have more CLV than others

```python
clv_data_cont_vars = clv_data[["Customer Lifetime Value","Income", "Monthly Premium Auto",
                               "Months Since Last Claim", "Months Since Policy Inception",
                               "Number of Open Complaints", "Number of Policies", "Total Claim Amount"]]
```

```python
plt.figure(figsize = [12, 8])

sns.heatmap(clv_data_cont_vars.corr(), annot=True)

plt.show()
```

higher monthly premiums and higher claims are associated with more expensive coverage

```python
plt.figure(figsize=[12, 8])

sns.scatterplot(x="Monthly Premium Auto", y="Total Claim Amount", hue='Coverage', data=clv_data)

plt.title("Claim Amount vs Monthly Premium Auto")
plt.show()
```

```python
plt.figure(figsize=[12, 8])

sns.scatterplot(x="Monthly Premium Auto", y="Customer Lifetime Value", hue="Coverage", data=clv_data)

plt.title("Monthly Premium Auto vs CLV")
plt.show()
```

Positive correlation can be seen.Similar patterns can be seen.
